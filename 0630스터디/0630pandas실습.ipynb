## 시리즈

```python
import pandas as pd

dict_data = {'a': 1, 'b': 2, 'c': 3} # 키 값이 그대로 index로 넘어오게 된다.

sr = pd.Series(dict_data) # 딕셔너리를 pandas 라이브러리의 Series()함수를 이용하여 딕셔너리를 함수의 매개변수로 전달

print(type(sr))
print()
print(sr)
```

```python
import pandas as pd

# 투플을 시리즈로 변환(index 옵션에 인덱스 이름을 지정)
tup_data = ('영인', '2010-05-01', '여', True)
sr = pd.Series(tup_data, index=['이름', '생년월일', '성별', '학생여부']) # index를 지정할 수도 있다
print(sr)
print()
# 원소를 1개 선택
print(sr[0]) # sr의 1 번째 원소를 선택 (정수형 위치 인덱스를 활용)
print(sr['이름']) # '이름' 라벨을 가진 원소를 선택 (인덱스 이름을 활용)

print('<<<원소를 선택한 경우>>>\n')
# 여러 개의 원소를 선택 (인덱스 리스트 활용)
print(sr[[1, 2]])
print(sr[['생년월일', '성별']])
print()

print('<<<인덱스 범위로 찾은 경우>>>\n')
# 여러 개의 원소를 선택 (인덱스 범위 지정)
print(sr[1 : 4]) 
print(sr['생년월일' : '학생여부'])
```

```python
import pandas as pd

# 행 인덱스/열 이름 지정하여, 데이터프레임 만들기
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']], 
index=['준서', '예은'], columns=['나이', '성별', '학교'])

# 데이터프레임 df 출력
print(df)
print("\n")

# 열 이름 중, '나이'를 '연령'으로, '성별'을 '남녀'로, '학교'를 '소속'으로 바꾸기
df = df.rename(columns = {'나이': '연령', '성별': '남녀', '학교': '소속'}) # 딕셔너리 형식을 사용하여 columns이냐 index냐의 차이에 따라 이름을 바꿀 수 있다.
# df의 행 인덱스 중에서, '준서'를 '학생1'로, '예은'을 '학생2'로 바꾸기
df = df.rename(index={'준서':'학생1', '예은':'학생2' })
# df 출력(변경 후)
print(df)
```

```python
df = df.drop('연령', axis=1) # inplace=True 하면 안됨;; 재정의 하는게 더 나을 수 도 있다. axis = 0이 디폴트 값(0은 행, 1은 열에서 검색)
df
```

# loc 와 iloc

- loc는 인덱스의 이름을 찾을 때

- iloc는 (앞의 i가 integer) 인덱스가 정수형일 때 인덱스의 위치를 찾을 때

```python
# DataFrame() 함수로 데이터프레임 변환. 변수 df에 저장
exam_data = {'수학' : [ 90, 80, 70], '영어' : [ 98, 89, 95],
'음악' : [ 85, 95, 100], '체육' : [ 100, 90, 90]}
df = pd.DataFrame(exam_data, index=['서준', '우현', '인아'])
print(df)
print('\n')

# 데이터프레임 df를 복제하여 변수 df2에 저장. df2의 1개 행(row)을 삭제
df2 = df[:]
df2 = df2.drop('우현') # 재 정의해서 덮어쓰는게 더 나은 것 같음 inplace사용 안하고!
print(df2)
print('\n')

# 데이터프레임 df를 복제하여 변수 df3에 저장. df3의 2개 행(row)을 삭제
df3 = df[:]
df3 = df3.drop(['우현', '인아'], axis=0) # 한 개 이상의 인덱스를 지울 때는 리스트로 묶기
print(df3)
```

```python
label1 = df.loc['서준']
position1 = df.iloc[0]

print(label1)
print(type(label1))
print()
print(position1)
```

# 여기부터 실습

```python
import pandas as pd

exam_data = {'이름': ['서준', '우현', '인아'],
             '수학' : [ 90, 80, 70],
             '영어' : [ 98, 89, 95],
             '음악' : [ 85, 95, 100],
             '체육' : [ 100, 90, 90]}
df = pd.DataFrame(exam_data)

df['총점'] = df['수학'] + df['영어'] + df['음악'] + df['체육']
df['평균'] = df['총점'] / 4
df
```

```python
df.loc[3]=['길동', 60, 70, 80, 90, None, None]
df['총점'] = df['수학'] + df['영어'] + df['음악'] + df['체육']
df['평균'] = df['총점'] / 4
df
```

```python
grades = []

for row in df['평균']:
    if row >= 90:
        grades.append('합격')
    else:
        grades.append('불합격')

df['합격여부'] = grades # 속성 값으로 리스트가 저장되면 

print(df.합격여부[0])
df
```

```python
def scholarship(row):
#     print(row)
    if row == '합격':
        return '장학생'
    else:
        return '비장학생'
    
df['합격여부'] = df.합격여부.apply(scholarship)
df
```

## 두 개의 데이터 프레임 합치기

```python
l1 = [{'name': 'John', 'job': "teacher"},
      {'name': 'Nate', 'job': "student"}, 
      {'name': 'Fred', 'job': "developer"}]

l2 = [{'name': 'Ed', 'job': "dentist"},
     {'name': 'Jack', 'job': "farmer"},
     {'name': 'Ted', 'job': "designer"}]

df1 = pd.DataFrame(l1, columns = ['name', 'job'])
df2 = pd.DataFrame(l2, columns = ['name', 'job'])
```

```python
df1
```

```python
df2
```

```python
merge_inner = pd.merge(df1, df2)

merge_inner # inner 조인 느낌 결과는 갑치는 부분이 없어서 출력이 안된 것 같다
```

```python
merge_outter = pd.merge(df1, df2, how = 'outer', on = 'name')
merge_outter
```
