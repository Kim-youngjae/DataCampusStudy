- age의 결측치를 생존여부, 성별에 따라 grouping 하고 각각의 평균을 결측치에 대입하라

```python
# Case 1

import pandas as pd

df = pd.read_csv('data/titanic.csv')
# df = df.set_index('PassengerId')

df_result = df.groupby(['Survived', 'Sex'])['Age'].mean()

for row in df.index:
    if str(df.loc[row, 'Age']) == 'nan':
        if (df.loc[row, 'Survived'] == 0) & (df.loc[row, 'Sex'] == 'male'):
            df.loc[row, 'Age'] = df_result[0]['male']
        elif (df.loc[row, 'Survived'] == 0) & (df.loc[row, 'Sex'] == 'female'):
            df.loc[row, 'Age'] = df_result[0]['female']
        if (df.loc[row, 'Survived'] == 1) & (df.loc[row, 'Sex'] == 'male'):
            df.loc[row, 'Age'] = df_result[1]['male']
        if (df.loc[row, 'Survived'] == 1) & (df.loc[row, 'Sex'] == 'female'):
            df.loc[row, 'Age'] = df_result[1]['female']

# df_result = df.groupby(['Survived', 'Sex'])['Age'].mean()
#             df의 Sex 와 Survived로 분류한

# print(df_result[0][0])


# df_result = df.groupby(['Sex', 'Survived'])['Age'].mean()
#             df의 Sex 와 Survived로 분류한

# print(df_result[0])

# df1 = df[(df['Survived'] == 0) & (df['Sex'] == 'male')]
# df2 = df[(df['Survived'] == 0) & (df['Sex'] == 'female')]
# df3 = df[(df['Survived'] == 1) & (df['Sex'] == 'male')]
# df4 = df[(df['Survived'] == 1) & (df['Sex'] == 'female')]


# df_result = pd.concat([df1, df2, df3, df4])


# df_result.info()
df = df[['PassengerId', 'Survived', 'Sex']]

# df_result.info()
df
```

```python
import pandas as pd

df = pd.read_csv('data/titanic.csv')

df1 = df[(df['Survived'] == 0) & (df['Sex'] == 'male')]
df2 = df[(df['Survived'] == 0) & (df['Sex'] == 'female')]
df3 = df[(df['Survived'] == 1) & (df['Sex'] == 'male')]
df4 = df[(df['Survived'] == 1) & (df['Sex'] == 'female')]

mean0 = df[(df['Survived'] == 0) & (df['Sex'] == 'male')]['Age'].mean()
mean1 = df[(df['Survived'] == 0) & (df['Sex'] == 'female')]['Age'].mean()
mean2 = df[(df['Survived'] == 1) & (df['Sex'] == 'male')]['Age'].mean()
mean3 = df[(df['Survived'] == 1) & (df['Sex'] == 'female')]['Age'].mean()

df1['Age'] = df1['Age'].fillna(mean0)
df2['Age'] = df2['Age'].fillna(mean1)
df3['Age'] = df3['Age'].fillna(mean2)
df4['Age'] = df4['Age'].fillna(mean3)

df_result = pd.concat([df1, df2, df3, df4])


# df_result.info()
df_result[['PassengerId', 'Survived', 'Sex']]
```

- 가상의 시나리오에서 1등석에 탑승한 승객은 구명보트 탑승시 우선 탑승권을 갖는다고 한다. 구조우선권을 기록하는 열을 추가하고 각 행에 구조 우선권을 가지면 o, 우선권이 없으면 x를 대입하라

```python
import pandas as pd

df = pd.read_csv('data/titanic.csv')

df = df[['PassengerId', 'Pclass']]

priority = []

for row in df['Pclass']:
    if row == 1:
        priority.append('o')
    else:
        priority.append('x')
        
df['priority'] = priority

df
```
